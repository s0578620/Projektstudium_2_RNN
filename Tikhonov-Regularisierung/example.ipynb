```python
import numpy as np
from scipy.linalg import inv
import matplotlib.pyplot as plt

# Daten
x_true = np.linspace(0, 2 * np.pi, 50)
y_true = np.sin(x_true)
y_noisy = y_true + 0.2 * np.random.randn(len(y_true))
A = np.vstack([x_true, np.ones(len(x_true))]).T

# Parameter
lmbda1 = 0
lmbda2 = 10.0
lmbda3 = 100.0


# λ1
I = np.identity(A.shape[1])
theta1 = inv(A.T @ A + lmbda1 * I) @ A.T @ y_noisy
y_reconstructed1 = A @ theta1

# λ2
theta2 = inv(A.T @ A + lmbda2 * I) @ A.T @ y_noisy
y_reconstructed2 = A @ theta2

# λ3
theta3 = inv(A.T @ A + lmbda3 * I) @ A.T @ y_noisy
y_reconstructed3 = A @ theta3

# Visualisierung
plt.plot(x_true, y_true, label='True Data', color='blue')
plt.scatter(x_true, y_noisy, label='Noisy Data', color='red', alpha=0.5)
plt.plot(x_true, y_reconstructed1, label=f'Reconstructed Data (λ={lmbda1})', color='green', linestyle='--')
plt.plot(x_true, y_reconstructed2, label=f'Reconstructed Data (λ={lmbda2})', color='orange', linestyle='--')
plt.plot(x_true, y_reconstructed3, label=f'Reconstructed Data (λ={lmbda3})', color='pink', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Tikhonov-Regularisierung (lineares Modell)')
plt.show()
```

```python
import numpy as np
from scipy.linalg import inv
import matplotlib.pyplot as plt

# Daten
x_true = np.linspace(0, 2 * np.pi, 50)
y_true = np.sin(x_true)
y_noisy = y_true + 0.2 * np.random.randn(len(y_true))
A = np.vstack([x_true**3, x_true**2, x_true, np.ones(len(x_true))]).T

# Parameter
lmbda1 = 0.1
lmbda2 = 10.0
lmbda3 = 100.0


# λ1
I = np.identity(A.shape[1])
theta1 = inv(A.T @ A + lmbda1 * I) @ A.T @ y_noisy
y_reconstructed1 = A @ theta1

# λ2
theta2 = inv(A.T @ A + lmbda2 * I) @ A.T @ y_noisy
y_reconstructed2 = A @ theta2

# λ3
theta3 = inv(A.T @ A + lmbda3 * I) @ A.T @ y_noisy
y_reconstructed3 = A @ theta3

# Visualisierung
plt.plot(x_true, y_true, label='True Data', color='blue')
plt.scatter(x_true, y_noisy, label='Noisy Data', color='red', alpha=0.5)
plt.plot(x_true, y_reconstructed1, label=f'Reconstructed Data (λ={lmbda1})', color='green', linestyle='--')
plt.plot(x_true, y_reconstructed2, label=f'Reconstructed Data (λ={lmbda2})', color='orange', linestyle='--')
plt.plot(x_true, y_reconstructed3, label=f'Reconstructed Data (λ={lmbda3})', color='pink', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Tikhonov-Regularisierung "(Polynommodell)')


plt.tight_layout()
plt.show()
```
